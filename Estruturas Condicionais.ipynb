# Estrutras Condicionais

Condicional

```python
# IF

if 4 > 3:
  print ("Resposta verdadeira")
```

```python
# IF / ELSE

num1 = 100
num2 = 20

if num1 > num2:
  print ("num1 é maior que num2")
else:
  print ("num2 é maior que num1")
```

```python
# IF / ELIF / ELSE

num3 = 50.1
num4 = 50.1

if num3 > num4:
  print ("num3 é maior!")
  print ("Valor de num3 é: ",num3)
elif num3 < num4:
  print ("num4 é maior!")
  print ("O valor de num4 é: ",num4)
else:
  print ("O valor de num3 e num4 é igual!",num3,"e",num4,)
```

Condicional com operadores lógicos (and, or, not)

```python
num5 = 100
num6 = 30
num7 = 50

if (num5 > num6) and (num5 > num7):
  print ("num5 é maior! Valor:",num5)
  print ("Outros valores:",num6,"e",num7)
else:
  print ("num5 não é maior.")
  print ("Outros valores:",num6,"e",num7)
```
